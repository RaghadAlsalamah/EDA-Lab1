```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("HR_Missing_Vslues.csv")
```

### Univariate Visualization 
* Plot two Charts
* Provide each charts with at least two insights 

```python
sns.countplot(data=df, y= "Absences",
              color = sns.color_palette()[0]);
plt.xticks(rotation=90);
```

#### Write your insghts here

- Most of the absences are 4 and 16 Absences
- All numbers of absences repeated from 5-10 times

```python
total = df['MaritalDesc'].value_counts()

plt.pie(total, labels=total.index, autopct='%1.2f%%',
       colors = sns.color_palette("pastel"));
```

#### Write your insghts here

- Single represent the largest percentage of employees
- Married represent the Second largest percentage of employees

### Bivariate Visualization 
* Plot two Charts
* Provide each charts with at least two insights 

```python
sns.lineplot(data=df, x="Absences", y="Salary");
```

#### Write your insghts here

- Every time there is an absence, the salary decreases
- 

```python
plt.bar(df['Department'], df['Salary'])
plt.xlabel('Department')
plt.ylabel('Salary')
```

#### Write your insghts here

- IT have higher salary 
- Software Engineering and Executive Office have similar salary

### Multivariate Visualization 
* Plot two Charts
* Provide each charts with at least two insights 

```python
sns.barplot( data=df, x="Department", y="Absences",  hue="Sex");
```

#### Write your insghts here

- The absence of the female more than the man
- Male admin offices has error estimation

```python
plt.figure(figsize=(10,10))
sns.ecdfplot(data=df,x='Salary',hue='Department',linewidth=3);
```

#### Write your insghts here

- All departments have an increase in salary
- Sales percentage increase is constant from 0.8 - 1.0
- IT have higher salary
